```python
# Import Pandas and Matplotlib dependencies

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Save each excel worksheet as a Pandas DataFrame

# Video 1

df1 = pd.read_excel('Resources/Video_1.xlsx', sheet_name=0)
df1_1 = pd.read_excel('Resources/Video_1.xlsx', sheet_name=1)
df1_2 = pd.read_excel('Resources/Video_1.xlsx', sheet_name=2)
df1_3 = pd.read_excel('Resources/Video_1.xlsx', sheet_name=3)
df1_4 = pd.read_excel('Resources/Video_1.xlsx', sheet_name=4)
df1_5 = pd.read_excel('Resources/Video_1.xlsx', sheet_name=5)

# Video 2

df2 = pd.read_excel('Resources/Video_2.xlsx', sheet_name=0)
df2_1 = pd.read_excel('Resources/Video_2.xlsx', sheet_name=1)
df2_2 = pd.read_excel('Resources/Video_2.xlsx', sheet_name=2)
df2_3 = pd.read_excel('Resources/Video_2.xlsx', sheet_name=3)
df2_4 = pd.read_excel('Resources/Video_2.xlsx', sheet_name=4)
df2_5 = pd.read_excel('Resources/Video_2.xlsx', sheet_name=5)
df2_6 = pd.read_excel('Resources/Video_2.xlsx', sheet_name=6)

# Video 3

df3 = pd.read_excel('Resources/Video_3.xlsx', sheet_name=0)
df3_1 = pd.read_excel('Resources/Video_3.xlsx', sheet_name=1)
df3_2 = pd.read_excel('Resources/Video_3.xlsx', sheet_name=2)
df3_3 = pd.read_excel('Resources/Video_3.xlsx', sheet_name=3)
df3_4 = pd.read_excel('Resources/Video_3.xlsx', sheet_name=4)
df3_5 = pd.read_excel('Resources/Video_3.xlsx', sheet_name=5)
```

```python
# Save updated DataFrames without "Total" row for better comparison

# Drop inconsistent rows (drop rows with 'Hashtag pages' and 'Sound pages' traffic source)

df1u = df1[1:].sort_values('Traffic source').drop([3,9]).reset_index(drop=True)
df2u = df2[1:].sort_values('Traffic source').reset_index(drop=True)
df3u = df3[1:].sort_values('Traffic source').drop(13).reset_index(drop=True)
```

```python
# Display the updated DataFrames

df1u
```

```python
df2u
```

```python
df3u
```

```python
# Use Matplotlib.pyplots to display comparitive views and save each figure to the Resources Folder

dfs = [df1u, df2u, df3u]
plt.yscale('log')
for x in dfs:
    plt.plot(x['Traffic source'], x['Views'])
plt.ylabel('Views')
plt.xticks(range(12),rotation=90)
plt.grid(alpha=.5)
plt.title('Views Per Video By Traffic Source')
plt.legend(['Video 1', 'Video 2', 'Video 3'])
plt.tight_layout()
plt.savefig('Resources/views_by_traffic_source.pdf')
plt.show()
```

```python
dfs = [df1u, df2u, df3u]
for x in dfs:
    plt.plot(x['Traffic source'], x['Average percentage viewed (%)'])
plt.ylabel('Average % Viewed')
plt.xticks(range(12),rotation=90)
plt.grid(alpha=.5)
plt.title('APV Per Video By Traffic Source')
plt.legend(['Video 1', 'Video 2', 'Video 3'], bbox_to_anchor=(1, 1), loc = 'lower right')
plt.tight_layout()
plt.savefig('Resources/APV_by_traffic_source.pdf')
plt.show()
```

```python
# With null-value rows dropped, view Impressions and CTR by traffic source

# Save figures to the Resources Folder

dfs = [df1u.dropna(), df2u.dropna(), df3u.dropna()]
plt.yscale('log')
for x in dfs:
    plt.plot(x['Traffic source'], x['Impressions'])
plt.ylabel('Impressions')
plt.xticks(range(6),rotation=90)
plt.grid(alpha=.5)
plt.title('Impressions Per Video By Traffic Source')
plt.legend(['Video 1', 'Video 2', 'Video 3'])
plt.tight_layout()
plt.savefig('Resources/impressions_by_traffic_source.pdf')
plt.show()
```

```python
dfs = [df1u.dropna(), df2u.dropna(), df3u.dropna()]
for x in dfs:
    plt.plot(x['Traffic source'], x['Impressions click-through rate (%)'])
plt.ylabel('Click Through Rate (%)')
plt.xticks(range(6),rotation=90)
plt.grid(alpha=.5)
plt.title('CTR Per Video By Traffic Source')
plt.legend(['Video 1', 'Video 2', 'Video 3'])
plt.tight_layout()
plt.savefig('Resources/CTR_by_traffic_source.pdf')
plt.show()
```
